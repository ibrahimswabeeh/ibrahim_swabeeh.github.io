```python
# PREPROCESSING PORR
# SPLIT THE DATA INTRO INPUT AND TARGET
X=df.iloc[:,1:-1]  
Y=df.iloc[:,-1:]
```

```python
# trace test split 
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest=train_test_split(X, Y, test_size=0.25, random_state=0)
print(Xtrain.shape, ytrain.shape)
print(Xtest.shape, ytest.shape)
```

 # KNN STEPS

-compute the distance btw test point vs all training samples
- (use any distance measures like Euclidien,Manhattan,minkowski)
- arrange the distance computed in ascending order
- select the k value
- vote for majority
- class having higher votes is the outcome
