```python
import numpy as np
import matplotlib.pyplot as plt

#single attributes
fs=1000
T=1/fs
f1=50
f2=100
t=np.arange(0,1,T)  # time samples
Y=np.sin(2*np.pi*f1*t)+(0.8*np.sin(2*np.pi*f2*t))
plt.plot(t, Y)
plt.title('signal plot')
plt.show()
```

```python
# compute the fft
Yfft=np.fft.fft(Y)
N=len(Y)  # no of samples(bin)
# to get actual freq value apply freqz function
Freqs=np.fft.fftfreq(N, d=T)
# fft computed also has positive and equivalent negative frequencies
plt.plot( Freqs[:N//2], np.abs(Yfft[:N//2]))
plt.title('fft plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# reproducible
np.random.seed(0)

fs = 1000
T = 1 / fs
t = np.arange(0, 1, T)

f1 = 50
f2 = 100

# noisy signal
noise = 0.5 * np.random.randn(len(t))
Y = np.sin(2*np.pi*f1*t) + 0.8*np.sin(2*np.pi*f2*t) + noise

# helper to compute single-sided FFT magnitude
def single_sided_fft(x, fs):
    N = len(x)
    X = np.fft.fft(x)
    mag = np.abs(X) / N
    freqs = np.fft.fftfreq(N, 1/fs)
    pos = freqs[:N//2]
    mag_pos = 2 * mag[:N//2]
    return pos, mag_pos

# plot time-domain raw
plt.figure(figsize=(10,3))
plt.plot(t, Y, lw=0.8)
plt.title("Noisy time signal")
plt.xlabel("Time (s)"); plt.ylabel("Amplitude")
plt.tight_layout()
```
