```python
# split data into input and target
X=df.iloc[:,:-1]
Y=df.iloc[:,-1:]
# Split the data into numeric and categorical
df_num=X.select_dtypes(include=np.number)
df_cat=X.select_dtypes(exclude=np.number)
```

```python
df_num
```

```python
df_cat.columns
```

```python
df_cat=df_cat.loc[:,['Sex', 'Diet', 'country', 'Continents', 'Hemisphere']]
df_cat
```

```python
df_cat.describe()
```

```python
# apply one hot encodingh onto categorial data
df_cat_en=pd.get_dummies(df_cat, drop_first=True,dtype='int')
df_cat_en
```

```python
# standardize the numerical data
#df_num.iloc[:,[0,1,2,8,11,12,13,14,15,17]]
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
df_num.iloc[:,[0,1,2,8,11,12,13,14,15,17]]=SS.fit_transform(df_num.iloc[:,[0,1,2,8,11,12,13,14,15,17]])
df_num
```

```python
df_num.head(2)
```

```python
# combine both catagorical and numerical data
X_cleaned=pd.concat([df_num, df_cat_en], axis=1)
X_cleaned
```

```python
# train test split the data
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest=train_test_split(X_cleaned, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
KNN=KNeighborsClassifier(n_neighbors=5, p=2)
KNN.fit(Xtrain, Ytrain)
```

```python
# testing
Ypredict=KNN.predict(Xtest)
cm=confusion_matrix(Ytest, Ypredict)
print(accuracy_score(Ytest, Ypredict))
cm
```

```python
# decision tree
DTC=DecisionTreeClassifier(criterion='entropy', random_state=0)

# train the model
DTC.fit(Xtrain, Ytrain)
Ypredict=DTC.predict(Xtest)
cm=confusion_matrix(Ytest, Ypredict)
accuracy_score(Ytest, Ypredict)
```

# hyperparameter tuning
- traul and error based

- grid and ramndom search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

# Define your training data and target values
# For example, you can load data from a dataset:
# from sklearn.datasets import load_iris
# data = load_iris()
# Xtrain = data.data
# Ytrain = data.target

# Or you can define them directly:
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load sample data
data = load_iris()
X = data.data
y = data.target

# Split the data into training and testing sets
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.3, random_state=42)

# Now proceed with GridSearchCV
Parameters={'n_estimators':[100,200,300], 'criterion':['entropy', 'gini'], 'max_depth':[4,6,8,10]}
Estimator=RandomForestClassifier(random_state=0)
RFC_grid=GridSearchCV(Estimator, Parameters, scoring='accuracy')
RFC_grid.fit(Xtrain, Ytrain)
print('Best Score=', RFC_grid.best_score_)
print('Best Parameters=', RFC_grid.best_params_)
```
