```python
# gennerate various signals
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,1,100)
S=np.sin(2*np.pi*3*t)
C=np.cos(2*np.pi*3*t)
D=np.tan(2*np.pi*3*t)
plt.subplot(3,1,1);
plt.plot(t,S,'r')
plt.title('sine wave')
plt.xlabel('time period')
plt.ylabel(' amplitude')

plt.subplot(3,1,2);
plt.plot(t,C,'k')
plt.title('cosine wave')

plt.subplot(3,1,3);
plt.plot(t,D,'g')
plt.title('tangent wave')
plt.xlabel(' timeperiod')
plt.ylabel(' amplitude')
plt.show()
```

```python
#generate various signals

import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,1,1000)
S1=np.sin(2*np.pi*3*t)
S2=np.sin(2*np.pi*10*t)
S3=np.sin(2*np.pi*30*t)
S4=np.sin(2*np.pi*50*t)
plt.subplot (2,2,1); plt.plot(t,S1)
plt.subplot (2,2,2); plt.plot(t,S2)
plt.subplot (2,2,3); plt.plot(t,S3)
plt.subplot (2,2,4); plt.plot(t,S4)
plt.show()

S=S1+S2+S3+S4  # signal mixing or addition in the time domain
plt.plot(t, S)
plt.show()

# convert to frq domain by applying fft

Sfft=np.fft.rfft(S) #converting time domain to freq domain
Flength=np.fft.rfftfreq(len(S), 1/1000)
plt.subplot (2,1,1); plt.plot(t, S); plt.title('time domain')
plt.subplot (2,1,2); plt.stem(Flength, np.abs(Sfft)); 
plt.title(' freq domain')
plt.show()

```
