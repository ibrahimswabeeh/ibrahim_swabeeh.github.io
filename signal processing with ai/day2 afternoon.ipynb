```python
import numpy as np
import matplotlib.pyplot as plt
fs=8000
t=np.arange(0,3, 1/fs)
y=np.sin(2*np.pi*1000*t)
plt.plot(t,y)

# to play it use ipython audio
from IPython.display import Audio
# to play normalize the negative freq part
y=y/np.max(np.abs(y))
Audio(y, rate=fs)
```
